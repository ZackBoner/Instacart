```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as ply
```

```python
orders = pd.read_csv("C:/Users/boner/Documents/Datasets/instacart.tar/instacart/instacart_2017_05_01/orders.csv")
products = pd.read_csv("C:/Users/boner/Documents/Datasets/instacart.tar/instacart/instacart_2017_05_01/products.csv")
order_products = pd.read_csv("C:/Users/boner/Documents/Datasets/instacart.tar/instacart/instacart_2017_05_01/order_products__prior.csv")
aisles = pd.read_csv("C:/Users/boner/Documents/Datasets/instacart.tar/instacart/instacart_2017_05_01/aisles.csv")
departments = pd.read_csv("C:/Users/boner/Documents/Datasets/instacart.tar/instacart/instacart_2017_05_01/departments.csv")
```

```python
# Only working with priors, so don't need the eval_set tag anymore
orders = orders[orders['eval_set']=='prior']
orders = orders.drop('eval_set',axis=1)
```
